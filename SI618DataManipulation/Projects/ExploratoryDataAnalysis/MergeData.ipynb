```python
import pandas as pd
import altair as alt
```

```python
df_eth = pd.read_csv(
    "Data/BTCUSDT.csv", index_col="TimeStamp"
).drop(["Unnamed: 0"], axis=1).rename(columns={
    "Close":"CloseETH",
    "TradeVolume":"TradeVolumeETH",
    "TradeCount":"TradeCountETH"
})
df_eth.index = pd.to_datetime(df_eth.index)
df_eth.sample(3)
```

```python
df_btc = pd.read_csv(
    "Data/BTCUSDT.csv", index_col="TimeStamp"
).drop(["Unnamed: 0"], axis=1)
df_btc.index = pd.to_datetime(df_btc.index)
df_btc.sample(3)
```

```python
ret = df_btc.merge(df_eth, left_index=True, right_index=True)
for field in ["Difficulty", "TransactionFee", "Transactions", "UniqueAddress", "MinerRevenue", "TotalMarketValue", "TotalHashRate"]:
    df = pd.read_csv("Data/{}.csv".format(field)).set_index(["TimeStamp"])
    df.index = pd.to_datetime(pd.to_datetime(df.index, format="%Y-%m-%d %H:%M:%S").date)
    df = df.groupby(df.index).mean()
    ret = ret.merge(df, left_index=True, right_index=True)
    print(ret.shape)
ret
```

```python
ret.to_csv("Data/Data.csv", index_label="TimeStamp")
```
