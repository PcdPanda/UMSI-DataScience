```python
import pandas as pd
import numpy as np
```

```python
# Q1
df = pd.read_csv("book_data.csv")
```

```python
# Q2
df = df.drop_duplicates(subset=["book_title"], keep="first")
"There are {} unique books".format(len(df["book_title"]))
```

```python
# Q3
df["author_num"] = df["book_authors"].apply(lambda x: len(x.split("|")))
"There are {} books with 2 authors".format(len(df[df["author_num"]==2]))
```

```python
# Q4
def get_page(book_pages):
    return int(book_pages.split(' ')[0])
df["pages"] = df["book_pages"].dropna().apply(get_page)
```

Some rows are droped because thy have nan on book pages

```python
# Q5
"The median number of pages is {}".format(df["pages"].median())
```

```python
"The minimum number of pages is {}, and the maximum number of page is {}".format(df["pages"].min(), df["pages"].max())
```

```python
# Q7
m1 = round(df[df["author_num"]==1]["pages"].mean(), 1)
m2 = round(df[df["author_num"]==2]["pages"].mean(), 1)
m3 = round(df[df["author_num"]==3]["pages"].mean(), 1)
"The average number of pages of books with one author is {}, for two authors is {}, for three authors is {}".format(m1, m2, m3)
```

**No, having more than 1 author doesn't result in a longer book on average**

```python
#Q8
"There are {} books have at least a rating of 4.5".format(len(df[df["book_rating"] >= 4.5]))
```

```python
# Q9
"There are {} books have at most a rating of 4".format(len(df[df["book_rating"] <= 4]))
```

```python
# Q10
df["discrete_rating"] = df["book_rating"].astype(int)
```

```python
# Q11
ans = list()
for i in range(1, 6):
    ans.append(round(df[df["discrete_rating"]==i]["author_num"].mean(), 1))
"For discretized rating 1, 2, 3, 4, 5, the average number of author is {}, {}, {}, {}, {} respectively".format(*ans)
```

```python
# Q12
ans = list()
for i in range(1, 6):
    ans.append(round(df[df["discrete_rating"]==i]["book_review_count"].mean(), 1))
"For discretized rating 1, 2, 3, 4, 5, the average number of review is {}, {}, {}, {}, {} respectively".format(*ans)
```

```python
# Q13
df["genres"] = df["genres"].apply(lambda x: str(x).split("|"))
genre_df = df.explode("genres", ignore_index=True)
```

```python
# Q14
for genre in set(genre_df["genres"]):
    avg = round(genre_df[genre_df["genres"]==genre]["book_rating"].mean(), 2)
    median = round(genre_df[genre_df["genres"]==genre]["book_rating"].median(), 2)
    print("The avg and median book rating for genre {} is {} and {}".format(genre, avg, median))
```

```python
# Q15
for genre in set(genre_df["genres"]):
    avg = round(genre_df[genre_df["genres"]==genre]["author_num"].mean(), 2)
    median = round(genre_df[genre_df["genres"]==genre]["author_num"].median(), 2)
    print("The avg and median author_num for genre {} is {} and {}".format(genre, avg, median))
```
